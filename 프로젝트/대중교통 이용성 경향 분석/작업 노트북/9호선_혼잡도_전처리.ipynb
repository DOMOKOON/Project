### 9호선 혼잡도 데이터는 여러 시트(평일,휴일,급행등)로 분리되어있어 한번에 합쳤다.

```python
import pandas as pd
file_path='/Users/parkjunhyeok/Desktop/codit/프로젝트/초급프로젝트/data/9호선 2022년 역별 시간별 혼잡도.xlsx'
sheets_dict = pd.read_excel(file_path, sheet_name=None)
```

```python
combined_df = pd.DataFrame()
for sheet_name, sheet_data in sheets_dict.items():
    sheet_data['시트'] = sheet_name  # 시트로 혼잡도가 나눠져있어서 시트 이름을 컬럼으로 추가
    combined_df = pd.concat([combined_df, sheet_data], ignore_index=True)
```

### 이후에 엑셀로 혼잡도데이터를 전처리했음(구분,시간대는 컬럼인데 시트마다 하나씩 들어가있어서 concat할때마다 하나의 행값이 구분과 시간대컬럼이 들어가서 삭제하고 이상치들을 제거했음)

```python
# 다시 다운받았음
c_df = pd.read_csv('./test9호선혼잡도.csv',encoding='utf-8-sig')
```

```python
c_df['sheet_name'].unique()
```

```python
c_df
```

```python
# 평일일반상선,하선을 평일일반으로 합치고
# 평일급행상선,하선을 평일급행으로 합치고
# 휴일일반상선,하선을 휴일일반으로 합치고
# 휴일급행상선,하선을 휴일급행으로 합쳤다.
# 이로써 평일,평일급행,휴일,휴일급행로 합쳐져있다.

# sheet_name 그룹화 정의
def classify_sheet_name(sheet_name):
    if sheet_name in ['상선일반(평일)', '하선일반(평일)']:
        return '평일 일반'
    elif sheet_name in ['상선급행(평일)', '하선급행(평일)']:
        return '평일 급행'
    elif sheet_name in ['상선일반(휴일)','하선일반(휴일)']:
        return '휴일 일반'
    else:
        return '휴일 급행'

# 새 그룹 열 추가
c_df['sheet_group'] = c_df['sheet_name'].apply(classify_sheet_name)

# 구분 및 sheet_group별 합계 계산
result = c_df.groupby(['구분', 'sheet_group']).sum().reset_index()
```

```python
result
```

### 1-8호선 혼잡도 데이터와 통일성을 가지기 위해 출근시간대,퇴근시간대를 나눠서 합치게 됨

```python
# 6시30분부터 9시까지 합치고 새로운 컬럼 추가 (오전 출근 시간대)
morning_time_cols = [
    '06:30~06:59', '07:00~07:29', '07:30~07:59',
    '08:00~08:29', '08:30~08:59'
]
result['오전출근시간대'] = result[morning_time_cols].sum(axis=1)

# 17시부터 19시30분까지 합치고 새로운 컬럼 추가 (오후 출근 시간대)
evening_time_cols = [
    '17:30~17:59', '18:00~18:29',
    '18:30~18:59', '19:00~19:29', '19:30~19:59'
]
result['오후출근시간대'] = result[evening_time_cols].sum(axis=1)

# 필요한 데이터 확인 및 저장
data = result[['구분', 'sheet_group', '오전출근시간대', '오후출근시간대', 'sheet_name']]
```

```python
data[data['구분']=='고속터미널']
# 컬럼이름이 sheet_group을 요일구분으로 바꿔줘
data.rename(columns={'sheet_group':'요일구분','오전출근시간대':'출근시간혼잡도','오후출근시간대':'퇴근시간혼잡도','구분':'출발역'}, inplace=True)
# sheet_name컬럼제거
data.drop(columns='sheet_name',inplace=True)
```

```python
# 호선 컬럼추가하고 9호선이기에 9로 채움
data['호선']='9'
```
