```python
re_df = pd.read_csv('./1-8호선_출근퇴근시간혼잡도.csv', encoding='cp949')
```

```python
re_df
#  출발역, 호선,역번호을 그룹화시켜서 요일구분에서 일요일,토요일은 합치고 휴일로 변경, 같은 역일때 출근시간혼잡도 , 퇴근시간혼잡도를 합산
```

### 일,토요일을 휴일로 합친이유는 9호선 데이터에서 휴일로 주말이나 공휴일을 통합했기때문에 

```python
# "일요일"과 "토요일" 데이터를 "휴일"로 합치기
re_df["요일구분"] = re_df["요일구분"].replace({"일요일": "휴일", "토요일": "휴일"})

# 같은 역번호, 호선, 요일구분, 출발역에 대해 출근시간혼잡도와 퇴근시간혼잡도를 합산
df_aggregated = (
    re_df.groupby(["역번호", "호선", "요일구분", "출발역"], as_index=False)
    .agg({"출근시간혼잡도": "sum", "퇴근시간혼잡도": "sum"})
)
```

```python
df_aggregated.head()
```

```python
df_aggregated.to_csv('./1-8호선최종혼잡도.csv', index=False, encoding='cp949')
```
